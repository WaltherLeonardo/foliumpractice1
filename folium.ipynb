```python
# nbi:hide_in
m= folium.Map(
    location=[-12.0983851,-77.0126965],
    tiles='Stamen Terrain',
    zoom_start=14
)

tooltip ='here I am!!'

folium.Marker(
    [-12.0983851,-77.0126965],
    popup='<i>here is the message</i>',
    tooltip=tooltip,
    icon=folium.Icon(icon='cloud')
).add_to(m)
folium.Marker(
    [-12.1183851,-77.0326965],
    popup='<b>and me??</b>',
    tooltip=tooltip,
    icon=folium.Icon(icon='info-sign', color='red')
).add_to(m)

m
```
